### Example 7:
In this example, we open the **prizes.json** file and iterate through the file to print only some of the values.<br>

We use **```pprint```** to make the output easier to read.

````python
import json
````

````python
from pprint import pprint
````

````python
with open("prizes.json", "r") as jsonfile:
    data = json.load(jsonfile)
````

````python
for row in data['prizes']:
    pprint(row['category'])
    pprint(row['year'])
````

### Example 4: 
In this example, we open a JSON file, use **```json.load```** to convert the content into a dictionary,and then use **```pprint```** to display the data from the dictionary in a human-readable format.<br>

To better understand what **```pprint```** does, you may want to open the raw JSON file on your computer to see how the data is stored in its original format.

````python
import json
````

````python
# We use pprint to display json/dictionary objects in a human-readable format
from pprint import pprint
````

````python
with open("prizes.json", "r") as jsonfile: # Use open to read the file
    # Use the json module with the load function
    # to read the content of the json file into a dictionary
    data = json.load(jsonfile) # Load the json content and serialize it
````

````python
print(type(data)) # Dict
````

````python
pprint(data) # Print the entire file content
````

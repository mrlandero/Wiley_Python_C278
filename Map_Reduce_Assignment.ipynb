```python
import csv
```

```python
def read_car_data(filepath,delimiter=","):
    dataset = list()
    with open(filepath) as f:
        csv_file = csv.DictReader(f,delimiter=",")
        for row in csv_file:
            dataset.append(row)
    return dataset
```

```python
read_car_data("car_data.csv")
```

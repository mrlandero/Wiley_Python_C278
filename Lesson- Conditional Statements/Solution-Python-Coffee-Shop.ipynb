## Description

```python
# Sample Answer for M1 Assignment/Self Assessment: Coffee Shop
# version 1 by David Albeck 2021-11-09

# A program that will calculate the cost of a custom cup of coffee at a gourmet coffee shop, 
#  based on the size of the cup, the type of coffee selected, and flavors that can be added to the coffee.

print("Welcome to Coffee.com!")

#  Ask the user what size cup they want, choosing between small, medium, and large.
size = input("Please choose a size of coffee cup (small/medium/large): ").lower()     #convert all inputs to lower case

#  Ask the user what kind of coffee they want, choosing between brewed, espresso, and cold brew.
brew = input("Please choose a brewing style (brewed/espresso/cold brew): ").lower()
             
#  Ask the user what flavoring they want, if any. Choices include hazelnut, vanilla, and caramel.
flavor = input("Please choose a secondary flavor (hazelnut/vanilla/caramel/none): ").lower()              
             
#  Calculate the price of the cup using the following values: (see assignment)

# variable "cost" will be a running tally
# initial value of cost, based on size:
if size == "small":
    cost = 2.00
elif size== "medium":
    cost = 3.00
elif size == "large":
    cost = 4.00
else:
    print("Sorry, couldn't understand what size you wanted.  Please start over.")
    exit()

# Add the charges for special brews, if any
if brew == "brewed":  # no extra charge
    cost += 0.00   # This line results in no change.  Program could be shorter --but confusing-- if we omitted this line.
elif brew == "espresso":
    cost += 0.50
elif brew == "cold brew":
    cost += 1.00
else:
    print("Sorry, couldn't understand what kind of brew you wanted. Please start over.")
    exit()
    
# Add the charges for extra flavoring, if any
if flavor != "none":  # Quick and dirty.  Charges the customer even if they request nonsensical flavors.
    cost += 0.50
    
# Display a statement that summarizes what the user ordered.
text = "You ordered a " + size + " cup of " + brew + " coffee"  
if flavor != "none":
    text += " with " + flavor + " syrup"
text += "."

print(text)

#  Display a total that corresponds to the price of the cup of coffee based on the user's choices.
print("Your coffee costs ${:.2f}".format(cost))

# Display a total that corresponds to the price of the cup of coffee with a 15% tip.
print("With a 15% tip, that'll be ${:.2f}".format(cost * 1.15))
```

### Example 1A:
In this example, we want to display the current date and time. This includes the following steps:<br>
    1. Import the ```datetime``` module.<br>
    2. Use the **datetime.now** class from the datetime module to calculate the current time.<br>
    3. Display the value generated by **datetime.datetime.now()**

````python
import datetime

# Assign the current date and time to the variable 'time_now'
time_now = datetime.datetime.now() # 
print(time_now)
````

### Example 1B:
Now let's see how to define a static date.<br>

We use the following syntax to define a date:<br>
```datetime.date(year,month,day)```

Each value must be an integer.<br>

In the following example, we generated Ada Lovelace's birthday, December 10,1815:

````python
import datetime

ada_bday = datetime.date(1815,12,10)
print(ada_bday)
````

### Example 1C:
We can use string functions to extract individual pieces of a datetime output.<br>
In the following example, we can see that the year, month, and day of the date are separated by a dash (-), and a space separates the date from the time. <br>
We first use the ```split``` function to separate the date and time, using the default space character.<br>
We then split the date into three separate values, specifying the dash as the separator.

````python
d = "2019-07-23 11:26:26.047342"
(d,t) = d.split()
(year,month,day) = d.split("-")
print(year)
print(month)
print(day)
print(t)
````

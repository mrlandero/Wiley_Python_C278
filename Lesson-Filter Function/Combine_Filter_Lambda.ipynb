### Example 2:
In this example, we import the data from a CSV file into a new list, where each row in the original file is a dictionary object in the list.<br>

Using a lambda function with the **```filter```** function, we then filter the data in the list to output only those records where the open price is lower than the closing price. The lambda function uses the key values "Open" and "Close" to create the comparison used by the **```filter```** function. The result is a **```filter```** object that includes only those items that meet the criteria in the lambda function.<br>

As a last step, we convert the **```filter```** object and display the filtered results from the dataset.

````python
def fromCSV(path,delimiter,quotechar):
    # Import the CSV module
    import csv
    # Any data we will read will be returned in a list
    data = list()
    # Open the file
    with open(path,newline="") as csvfile:
        # Access the contents of the file
        filecontent = csv.DictReader(csvfile,delimiter=delimiter,quotechar=quotechar)
        # Iterate through the rows
        for row in filecontent:
            # Save the rows in the data list. Each row is a dictionary
            data.append(row)
    return data
````

````python
data = fromCSV(path="stocks.csv",delimiter=",",quotechar="|")
````

````python
# Filter all elements in data where the open price is lower than the closing price
data_filtered = filter(lambda x: x["Open"] < x["Close"], data)
print(type(data_filtered))
````

````python
# Convert the filter object into a list
data_filtered_list = list(data_filtered)
# Display each element in the filtered list
for row in data_filtered_list:
    print(row)
````

````python
# Filter all elements in data where the high price is lower than the closing price
data2 = fromCSV(path="stocks.csv",delimiter=",",quotechar="|")
data_filtered2 = filter(lambda s: s['High'] < s['Close'], data2)
print(type(data_filtered2))
````

````python
# Convert the filter object into a list
data_filtered_list2 = list(data_filtered2)
# Display each element in the filtered list
for row in data_filtered_list2:
    print(row)
````

### Example 6: 
This example uses the same **```SeqDiv3```** class we created earlier, with a **```StopIteration```** on the value of 20.<br>

Instead of using a series of **```next```** statements to output the values generated by the class, we can use a **```for```** loop, which stops automatically when the current value reaches 20.

````python
class SeqDiv3:
    def __iter__(self):
        self.a=3
        return self
    def __next__(self):
        if(self.a<20):
            x = self.a
            self.a +=3
            return x
        raise StopIteration
````

````python
numbers = SeqDiv3()
for number in numbers: # This will stop at 18 because we implemented StopIteration
    print(number)
````
